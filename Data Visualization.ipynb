Import necessary libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the dataset

```python
iris = sns.load_dataset('iris')
```

```python
print(iris.head())
```

Inspect the dataset

```python
# Print the shape of the dataset (rows and columns)
print(f"Shape of dataset: {iris.shape}")
```

```python
# Print the column names
print(f"Columns in dataset: {iris.columns}")
```

```python
# Get summary statistics of the dataset
print(iris.describe())
```

```python
# Get information about dataset types and null values
print(iris.info())
```

Visualize the dataset

Scatter Plot

```python
# Scatter plot showing relationships between features
sns.pairplot(iris)
plt.show()
```

Histograms

```python
# Plot histograms for all numeric columns
iris.hist(bins=20, figsize=(10, 8))
plt.show()
```

Box Plots

```python
# Boxplot for each feature
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris)
plt.show()
```
